```python
# Import data from csv file

import numpy as np
import csv
import datetime as dt

with open('grades.csv', 'r', encoding='utf-8-sig') as f:
    reader = csv.reader(f)
    data = np.array(list(reader))

x1 = [float(v) for v in data[:,0]]
x2 = [float(v) for v in data[:,1]]
X = np.transpose(np.array([x1, x2]))
y = np.array([float(v) for v in data[:, 2]])
```

```python
# Histograms

import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
ax1.hist(x1)
ax1.set_title('Study time')
ax2.hist(x2)
ax2.set_title('Sleep time')
ax3.hist(y)
ax3.set_title('Grade')
```
